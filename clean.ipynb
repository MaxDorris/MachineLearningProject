```python
# convert crime lat,lng to nearest zipcode - input is an array of lat, long
def get_zip(coords):
    coord_dists = zipcode_df
    dists = []
    for i in coord_dists.values:
        dists.append(pgeocode.haversine_distance(np.array(coords), np.array([[i[0],i[1]]]))[0])

    coord_dists['dist'] = dists
    return coord_dists.loc[coord_dists['dist'] == min(coord_dists['dist'])].index
```
